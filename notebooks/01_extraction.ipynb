# 01 Data Extraction & Merging

In this notebook, we load the two raw datasets, perform a preliminary analysis of their structure, and merge them into a single unified dataset for further cleaning.

## Objectives:
- Load `game_data_all.csv` and `games-release-ALL.csv`.
- Identify the common key for merging.
- Merge the datasets and save the result to `data/merged/`.

```python
import pandas as pd
import os

raw_path = '../data/raw/'
merged_path = '../data/merged/'

if not os.path.exists(merged_path):
    os.makedirs(merged_path)

file1 = os.path.join(raw_path, 'game_data_all.csv')
file2 = os.path.join(raw_path, 'games-release-ALL.csv')

print(f"Loading {file1}...")
df1 = pd.read_csv(file1)

print(f"Loading {file2}...")
df2 = pd.read_csv(file2)

print(f"Shape of df1: {df1.shape}")
print(f"Shape of df2: {df2.shape}")
```

## Column Analysis
Let's look at the columns of both datasets to decide on the merge key.

```python
print("Columns in df1:", df1.columns.tolist())
print("Columns in df2:", df2.columns.tolist())
```

## Merging Datasets
Both datasets have a `link` column which represents the unique Steam App path. This is the most reliable key for merging.

We will use an **outer merge** to ensure we capture all unique games from both sources, though they appear to have significant overlap.

```python

df_merged = pd.merge(df1, df2, on='link', how='outer', suffixes=('', '_drop'))

print(f"Shape after merge: {df_merged.shape}")

cols_to_drop = [c for c in df_merged.columns if c.endswith('_drop')]
df_merged.drop(columns=cols_to_drop, inplace=True)

print(f"Final Shape: {df_merged.shape}")
```

## Export Merged Data
Saving the merged dataset to `data/merged/merged_game_data.csv`.

```python
output_file = os.path.join(merged_path, 'merged_game_data.csv')
df_merged.to_csv(output_file, index=False)
print(f"Merged data saved to: {output_file}")
```
